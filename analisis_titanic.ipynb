```python
# load data & explorary awal

import pandas as pd

#baca file titanic
data = pd.read_csv("titanic.csv")

#lihat lima baris pertama
data.head()

#info kolom dan tipe data
data.info()

#satistik dasar
data.describe()
```

```python
# Deteksi masalah data

#cek jumlah data kosong 
data.isnull().sum()

#cek data duplicat
data.duplicated().sum()
```

```python
# Imputasi data

# Isi nilai kosong di kolom age dengan rata-rata
data['age'].fillna(data['age'].mean(), inplace=True)

# Cek ulang apakah sudah bersih
data.isnull().sum()
```
